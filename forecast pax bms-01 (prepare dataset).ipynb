## Combine Data

```python
frames = [data_april, data_mei, data_juni, data_juli]
combine_data = pd.concat(frames)
```

```python
df = combine_data[['id_seq', 'route_info', 'shelter_id', 'created_on']]
df
```

```python
df = df[df.route_info != 'RUTE_TESTING']
df
```

### Create shelter name column

```python
shelter_name = []
for i in range(len(df)):
    if df['shelter_id'].values[i] == 1:
        shelter_name.append('T1')
    elif df['shelter_id'].values[i] == 4:
        shelter_name.append('T2')
    elif df['shelter_id'].values[i] == 5:
        shelter_name.append('T3')
print(len(shelter_name))
```

```python
df['shelter_name'] = shelter_name
df
```

```python
df.info()
```

## Create Hourly Data

```python
start = datetime.datetime(2023, 4, 1)
end = datetime.datetime(2023, 7, 20)
step = datetime.timedelta(hours=1)

date = []

while start <= end:
    date.append(start.strftime('%Y-%m-%d %H:%M:%S'))
    start += step
date = pd.to_datetime(pd.Series(date), format='%Y-%m-%d %H:%M:%S')

date_time = []
for i in range(len(date[:])):
    date_time.append(date[i].strftime("%Y-%m-%d %H:%M:%S"))
print(date_time[:5]);print(date_time[-5:])
print(len(date_time))
```

```python
def get_pax_bms(terminal):
    num_pax = []
    for i in range(len(date_time[:-1])):
        time2 = datetime.datetime.strptime(str(date_time[i+1]), '%Y-%m-%d %H:%M:%S') - datetime.timedelta(minutes=1)
        temp = df[df['created_on'] > date_time[i]]
        temp = temp[temp['created_on'] < time2]
        temp_ = temp[temp['shelter_name'] == f'T{terminal}']
        num_pax.append(len(temp_))
    print(f'length data pax bms terminal {terminal}:',len(num_pax))
    return(num_pax)
```

```python
pax_bms_t1 = get_pax_bms(terminal=1)
pax_bms_t2 = get_pax_bms(terminal=2)
pax_bms_t3 = get_pax_bms(terminal=3)
```

```python
data_bms = pd.DataFrame()
data_bms["DATE"] = pd.to_datetime(date_time[:-1], errors = 'coerce')
data_bms["PAX_BMS_T1"] = pax_bms_t1
data_bms["PAX_BMS_T2"] = pax_bms_t2
data_bms["PAX_BMS_T3"] = pax_bms_t3

data_bms
```

## Visualisasi Data

```python
plt.subplots(figsize=(25, 5))
ax = sns.lineplot(x="DATE", y="PAX_BMS_T1", data=data_bms, 
                    label="Pax BMS T1", marker='.', color='deepskyblue')
plt.xlabel('Waktu'); plt.ylabel('Jumlah Penumpang')
plt.title("Data Jumlah Pax BMS per Terminal 1 CGK\nPada 1 April - 19 Juli 2023\n", fontsize = 15)
plt.grid()
plt.legend(loc='upper left')
```
